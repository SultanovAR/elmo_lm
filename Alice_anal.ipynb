```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,4))
titles = ['num of examples', 'length of sentence', 'uniq tokens']
for num, i in enumerate(imdb):
    plt.subplot(1, 3, num+1)
    plt.title(titles[num])
    plt.bar(list(imdb[i].keys()), list(imdb[i].values()))
plt.tight_layout()
```
